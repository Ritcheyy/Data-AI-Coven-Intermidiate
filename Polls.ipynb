```python
import pandas as pd
import csv
```

```python
data_info = []
for page in range(1,3):
    # Website Url
    page_url = "http://integrity.ng/index.php/units/browse/{}".format(page)
    
    # Append each page data in an array
    data_info.append(pd.read_html(page_url)[0])
```

```python
df = pd.concat(data_info)
```
